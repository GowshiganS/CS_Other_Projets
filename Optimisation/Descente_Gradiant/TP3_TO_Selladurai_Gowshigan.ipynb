## Exercice 1 : Modélisation mathématique
### 1. Calculer le gradient ∇f (x, y) de la fonction donnée
Faisons la dérivée partielle par rapport à x:
$$
\frac{\partial f(x, y)}{\partial x} = 2ax + cy + d
$$
puis par rapport à y :
$$
\frac{\partial f(x, y)}{\partial y} = -2by + cx + e

### 2. Montrer que la fonction est convexe en vérifiant que la matrice hessienne H est définie positive.
$$
f(x, y) = ax^2 + by^2 + cxy + dx + ey + f
$$

#### Étape 1 : Dérivées premières

- **Dérivée partielle de \( f \) par rapport à \( x \)** :
  $$
  \frac{\partial f}{\partial x} = 2ax + cy + d
  $$

- **Dérivée partielle de \( f \) par rapport à \( y \)** :
  $$
  \frac{\partial f}{\partial y} = 2by + cx + e
  $$

#### Étape 2 : Dérivées secondes

- **Dérivée seconde par rapport à \( x \)** :
  $$
  \frac{\partial^2 f}{\partial x^2} = 2a
  $$

- **Dérivée seconde par rapport à \( y \)** :
  $$
  \frac{\partial^2 f}{\partial y^2} = 2b
  $$

  
- **Dérivée croisée par rapport à \( x \) et \( y \)** :
  $$ \frac{\partial f}{\partial x}$$ par rapport à \(y\) :
  $$
  \frac{\partial f}{\partial x} = 2ax + cy + d
  $$
  En dérivant par rapport à \(y\), on obtient :
  $$
  \frac{\partial^2 f}{\partial x \partial y} = c
  $$

- **Dérivée croisée par rapport à \( y \) et \( x \)** :
  $$\frac{\partial f}{\partial y} $$ par rapport à \(x\) :
  $$
  \frac{\partial f}{\partial y} = 2by + cx + e
  $$
  En dérivant par rapport à \(x\), on obtient :
  $$
  \frac{\partial^2 f}{\partial y \partial x} = c
  $$

#### Étape 3 : Matrice hessienne

La matrice hessienne \( H(x, y) \) :

$$
H(x, y) = 
\begin{pmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{pmatrix}
= 
\begin{pmatrix}
2a & c \\
c & 2b
\end{pmatrix}
$$

### Étape 4 : Conditions pour la convexité

La fonction est convexe si la matrice hessiene est postif, avec pour conditions

1. Les éléments diagonaux positifs :
   $$
   2a > 0 \quad \text{et} \quad 2b > 0
   $$
   Ce qui équivaut à \( a > 0 \) et \( b > 0 \).

2. Le déterminant strictement positif :
   $$
   \text{Dét}(H) = (2a)(2b) - c^2 = 4ab - c^2 > 0
   $$

Donc, la fonction est convexe si :
$$
a > 0, \quad b > 0, \quad \text{et} \quad 4ab - c^2 > 0.
$$






### 3.Écrire la règle de mise à jour pour les variables x et y.

#### Règle de mise à jour

La règle de mise à jour est donnée par :
$$
\mathbf{x}_{k+1} = \mathbf{x}_k - \eta \nabla f(\mathbf{x}_k)
$$

Le gradient est :

$$
\nabla f(x, y) = 
\begin{pmatrix}
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial y}
\end{pmatrix}
$$

Avec :
- **Dérivée partielle par rapport à \(x\)** :
  $$
  \frac{\partial f}{\partial x} = 2ax + cy + d
  $$

- **Dérivée partielle par rapport à \(y\)** :
  $$
  \frac{\partial f}{\partial y} = 2by + cx + e
  $$

Ainsi :

$$
\nabla f(x, y) = 
\begin{pmatrix}
2ax + cy + d \\
2by + cx + e
\end{pmatrix}
$$

---

#### Règle de mise à jour explicite

En appliquant cette règle, les mises à jour des variables \(x\) et \(y\) deviennent :

- **Mise à jour de \(x\)** :
  $$
  x_{k+1} = x_k - \eta (2ax_k + cy_k + d)
  $$

- **Mise à jour de \(y\)** :
  $$
  y_{k+1} = y_k - \eta (2by_k + cx_k + e)
  $$

## Exercice 2 : Résolution manuelle 


$$
f(x, y) = x^2 + y^2 - 2x - 2y
$$

On va faire les dérivées partielles de la fonction par rapport à \( x \) et \( y \).

---

### Étape 1 : Calcul des dérivées partielles

- **Dérivée partielle de \( f \) par rapport à \( x \)** :
  $$
  \frac{\partial f}{\partial x} = 2x - 2
  $$

- **Dérivée partielle de \( f \) par rapport à \( y \)** :
  $$
  \frac{\partial f}{\partial y} = 2y - 2

---

### Étape 2 : Mise en place de la méthode de descente de gradient

La formule de mise à jour de la descente de gradient est la suivante :

$$
x_{k+1} = x_k - \eta \frac{\partial f}{\partial x}(x_k, y_k)
$$
$$
y_{k+1} = y_k - \eta \frac{\partial f}{\partial y}(x_k, y_k)
$$

Avec les conditions initiales :
$$
x_0 = 0, \quad y_0 = 0, \quad \eta = 0.1
$$


### Étape 3 : Itérations de la descente de gradient

#### **Itération 1**  :

- **Calcul des gradients** :
  $$
  \frac{\partial f}{\partial x}(0, 0) = 2(0) - 2 = -2
  $$
  $$
  \frac{\partial f}{\partial y}(0, 0) = 2(0) - 2 = -2
  $$

- **Mise à jour des valeurs** :
  $$
  x_1 = x_0 - \eta \frac{\partial f}{\partial x}(0, 0) = 0 - 0.1(-2) = 0.2
  $$
  $$
  y_1 = y_0 - \eta \frac{\partial f}{\partial y}(0, 0) = 0 - 0.1(-2) = 0.2
  $$

Donc, après la première itération :
$$
x_1 = 0.2, \quad y_1 = 0.2
$$

---

#### **Itération 2**:

- **Calcul des gradients** :
  $$
  \frac{\partial f}{\partial x}(0.2, 0.2) = 2(0.2) - 2 = -1.6
  $$
  $$
  \frac{\partial f}{\partial y}(0.2, 0.2) = 2(0.2) - 2 = -1.6
  $$

- **Mise à jour des valeurs** :
  $$
  x_2 = x_1 - \eta \frac{\partial f}{\partial x}(0.2, 0.2) = 0.2 - 0.1(-1.6) = 0.36
  $$
  $$
  y_2 = y_1 - \eta \frac{\partial f}{\partial y}(0.2, 0.2) = 0.2 - 0.1(-1.6) = 0.36
  $$

Donc, après la deuxième itération :
$$
x_2 = 0.36, \quad y_2 = 0.36
$$

---

#### **Itération 3** :

- **Calcul des gradients** :
  $$
  \frac{\partial f}{\partial x}(0.36, 0.36) = 2(0.36) - 2 = -1.28
  $$
  $$
  \frac{\partial f}{\partial y}(0.36, 0.36) = 2(0.36) - 2 = -1.28

  $$

x_3 = 0.488, \quad y_3 = 0.488

## Exercice 3 : Implémentation Python

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x, y):
    return x**2 + y**2 - 2*x - 2*y

# Fonction pour calculer le gradient avec des différences finies
def gradient(f, x, y, h=1e-5):
    # Calcul des dérivées partielles
    # h permet de d'approximer les dérivées
    
    # df_dx est l'approximation de la dérivée partielle par rapport à x
    df_dx = (f(x + h, y) - f(x, y))/ h
    # df_dy est l'approximation de la dérivée partielle par rapport à y
    df_dy = (f(x, y + h) - f(x, y)) / h
    # on retourne sous forme de tableau NumPy contenant les dérivées partielles par rapport à x et y.
    return np.array([df_dx, df_dy])

```

```python

#l'algorithme de descente de gradient
def gradient_descent(f,grad, x_init, y_init, eta=0.1, tol=1e-5, max_iter=1000):
    x, y = x_init, y_init
    trajectory = [(x, y)]  # Liste pour stocker les positions des itérations

    for i in range(int(max_iter)):
        grad_x, grad_y = gradient(f,x, y)
        # calcul des normes du gradient
        norm_grad = np.linalg.norm([grad_x, grad_y])

        # Mise à jour des valeurs de x et y
        x = x - eta * grad_x
        y = y - eta * grad_y

        # Sauvegarder les valeurs de x et y
        trajectory.append((x, y))

        # Afficher les valeurs à chaque itération
        print(f"Itération {i+1}: x = {x}, y = {y}, f(x, y) = {f(x, y)}")

        # Vérifier la condition d'arrêt
        if norm_grad < tol:
            print(f"Convergence atteinte après {i+1} itérations.")
            break

    return x, y, trajectory

# Initialisation des valeurs
x_init, y_init = 0, 0  # Point de départ
eta = 0.1  # Taux d'apprentissage
tol = 1e-5  # Tolérance pour la norme du gradient

# 
x_min, y_min, trajectory = gradient_descent(f,gradient, x_init, y_init, eta, tol)

# Afficher les résultats
print(f"\nMinimum atteint à : x = {x_min}, y = {y_min}, f(x, y) = {f(x_min, y_min)}")

# Visualisation des contours de f(x, y) et des trajectoires
x_vals = np.linspace(-1, 3, 400)
y_vals = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = f(X, Y)

# Tracer les contours de la fonction
CS = plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Extraire les trajectoires des itérations
trajectory = np.array(trajectory)

# Tracer la trajectoire de la descente de gradient
plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', color='r', markersize=5, label="Trajectoire")


plt.title("Trajectoire de la descente de gradient")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.colorbar(CS, label="f(x, y)")

# Afficher le graphique
plt.show()
```

## Conclusion:

### Avantages et limitations : 
#### Avantages:
- Très utilisé dans le domaine du datascience
- Flexiblilité d'utilisation sur une large gamme de fonctions
#### limitations:
- Mauvais choix du taux d'apprentissage peut rendre l'algorithme très lent ou infini ( pour un eta trop petit ou trop grand)
- L'algorithme risque d'être bloqué dans un point de minima local au lieu d'atteindre le minima global
- Oscillations entre des minima locaux peut ralentir la convergeance 
### Améliorations possibles :
#### l’utilisation du moment:
- Permet d'utiliser les valeurs précédentes du chemin du gradient afin  d'avoir une direction stable et réduire les oscillations et ainsi accélérer la convergeance
#### d’un taux d’apprentissage adaptatif:
- On ajuste le taux d'apprentissage selon les itérations ou les paramètres afin d'accélérer la convergeance
