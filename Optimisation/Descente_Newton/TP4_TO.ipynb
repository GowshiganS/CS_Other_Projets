$$

#### Étape 1 : Dérivées premières

- **Dérivée partielle de \( f \) par rapport à \( x \)** :
  $$

- **Dérivée partielle de \( f \) par rapport à \( y \)** :
  $$

  $$

#### Étape 2 : Dérivées secondes

- **Dérivée seconde par rapport à \( x \)** :
  $$
  \frac{\partial^2 f}{\partial x^2} = 2a
  $$

- **Dérivée seconde par rapport à \( y \)** :
  $$
  \frac{\partial^2 f}{\partial y^2} = 2b
  $$

  
- **Dérivée croisée par rapport à \( x \) et \( y \)** :
  $$ \frac{\partial f}{\partial x}$$ par rapport à \(y\) :
  $$
  \frac{\partial f}{\partial x} = 2ax + cy + d
  $$
  En dérivant par rapport à \(y\), on obtient :
  $$
  \frac{\partial^2 f}{\partial x \partial y} = c
  $$

- **Dérivée croisée par rapport à \( y \) et \( x \)** :
  $$\frac{\partial f}{\partial y} $$ par rapport à \(x\) :
  $$
  \frac{\partial f}{\partial y} = 2by + cx + e
  $$
  En dérivant par rapport à \(x\), on obtient :
  $$
  \frac{\partial^2 f}{\partial y \partial x} = c
  $$

#### Étape 3 : Matrice hessienne

La matrice hessienne \( H(x, y) \) :

$$
H(x, y) = 
\begin{pmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{pmatrix}
= 
\begin{pmatrix}
2a & c \\
c & 2b
\end{pmatrix}
$$

#### Étape 4 : Conditions pour la convexité

La fonction est convexe si la matrice hessiene est postif, avec pour conditions

1. Les éléments diagonaux positifs :
   $$
   2a > 0 \quad \text{et} \quad 2b > 0
   $$
   Ce qui équivaut à \( a > 0 \) et \( b > 0 \).

2. Le déterminant strictement positif :
   $$
   \text{Dét}(H) = (2a)(2b) - c^2 = 4ab - c^2 > 0
   $$

Donc, la fonction est convexe si :
$$
a > 0, \quad b > 0, \quad \text{et} \quad 4ab - c^2 > 0.
$$

## Exercice 3 : Implémentation Python 

```python
import numpy as np
import matplotlib.pyplot as plt


def f(x, y):
    return x**2 + y**2 - 2*x - 2*y

# Fonction pour calculer le gradient avec des différences finies
def gradient(f, x, y, h=1e-5):
    # Calcul des dérivées partielles
    # h permet de d'approximer les dérivées
    
    # df_dx est l'approximation de la dérivée partielle par rapport à x
    df_dx = (f(x + h, y) - f(x, y))/ h
    # df_dy est l'approximation de la dérivée partielle par rapport à y
    df_dy = (f(x, y + h) - f(x, y)) / h
    # on retourne sous forme de tableau NumPy contenant les dérivées partielles par rapport à x et y.
    return np.array([df_dx, df_dy])

# Fonction pour approximer de la matrice hessienne par différences finies."""
def hessienne(f, x, y, h=1e-5):
   
    # Calcul des dérivées secondes
    d2f_dx2 = (f(x + h, y) - 2*f(x, y) + f(x - h, y)) / h**2
    d2f_dy2 = (f(x, y + h) - 2*f(x, y) + f(x, y - h)) / h**2
    d2f_dxdy = (f(x + h, y + h) - f(x + h, y - h) - f(x - h, y + h) + f(x - h, y - h)) / (4 * h**2)
    # Construire la matrice hessienne
    return np.array([[d2f_dx2, d2f_dxdy], [d2f_dxdy, d2f_dy2]])
```

```python
def newton(f,x0, y0, tol=1e-6, max_iter=100):
 
    x, y = x0, y0
    trajectory = [(x, y)]
    for _ in range(max_iter):
        grad = gradient(f, x, y)
        hess = hessienne(f, x, y)
        try:
            hess_inv = np.linalg.inv(hess)
        except np.linalg.LinAlgError:
            print("Matrice hessienne non inversible.")
            break
        step = np.dot(hess_inv, grad)
        x, y = x - step[0], y - step[1]
        trajectory.append((x, y))
        
        # Condition d'arrêt : norme du gradient
        if np.linalg.norm(grad) < tol:
            break
    return x, y, f(x, y), trajectory

def coutour(trajectory):
    
    x_vals = np.linspace(-1, 3, 100)
    y_vals = np.linspace(-1, 3, 100)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = f(X, Y)
    
    # Plot contours
    plt.figure(figsize=(8, 6))
    plt.contour(X, Y, Z, levels=20, cmap='viridis')
    plt.colorbar(label='f(x, y)')
    
    # Plot trajectoire
    trajectory = np.array(trajectory)
    plt.plot(trajectory[:, 0], trajectory[:, 1], 'ro-', label='Trajectory')
    plt.scatter(trajectory[:, 0], trajectory[:, 1], color='red')
    
    plt.title("Methode Newton trajectoire")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

# Initial point
x0, y0 = 0, 0

# Lancer la fonction newton
x_opt, y_opt, f_opt, trajectory = newton(f,x0, y0)

# Affichage des résultats
print(f"point Optimale: x = {x_opt}, y = {y_opt}")
print(f"value Optimale of f(x, y): {f_opt}")

# Plot du trajectoire
coutour(trajectory)
```

```python
import math
def g(x,y):
    return np.cos(x)+np.sin(y)+x**2+y**2
  
# Initialisation des valeurs
x_init, y_init = 0, 0  # Point de départ

# Faire la descente
x_min, y_min, g_opt, trajectory = newton(g,x_init, y_init)

# Afficher les résultats
print(f"\nMinimum atteint à : x = {x_min}, y = {y_min}, g(x, y) = {g(x_min, y_min)}")

# Visualisation des contours de g(x, y) et des trajectoires
x_vals = np.linspace(-1, 3, 400)
y_vals = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = g(X, Y)

# Tracer les contours de la fonction
CS = plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Extraire les trajectoires des itérations
trajectory = np.array(trajectory)

# Tracer la trajectoire de la descente de gradient
plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', color='r', markersize=5, label="Trajectoire")


plt.title("Trajectoire de la descente de gradient")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.colorbar(CS, label="g(x, y)")

# Afficher le graphique
plt.show()
```

```python
def gradient_descent(f,grad, x_init, y_init, eta=0.1, tol=1e-5, max_iter=1000):
    x, y = x_init, y_init
    trajectory = [(x, y)]  # Liste pour stocker les positions des itérations

    for i in range(int(max_iter)):
        grad_x, grad_y = gradient(f,x, y)
        # calcul des normes du gradient
        norm_grad = np.linalg.norm([grad_x, grad_y])

        # Mise à jour des valeurs de x et y
        x = x - eta * grad_x
        y = y - eta * grad_y

        # Sauvegarder les valeurs de x et y
        trajectory.append((x, y))

        # Afficher les valeurs à chaque itération
        print(f"Itération {i+1}: x = {x}, y = {y}, f(x, y) = {f(x, y)}")

        # Vérifier la condition d'arrêt
        if norm_grad < tol:
            print(f"Convergence atteinte après {i+1} itérations.")
            break

    return x, y, trajectory

# Initialisation des valeurs
x_init, y_init = 0, 0  # Point de départ
eta = 0.5  # Taux d'apprentissage
tol = 1e-5  # Tolérance pour la norme du gradient

# 
x_min, y_min, trajectory = gradient_descent(g,gradient, x_init, y_init, eta, tol)

# Afficher les résultats
print(f"\nMinimum atteint à : x = {x_min}, y = {y_min}, f(x, y) = {f(x_min, y_min)}")

# Visualisation des contours de f(x, y) et des trajectoires
x_vals = np.linspace(-1, 3, 400)
y_vals = np.linspace(-1, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = g(X, Y)

# Tracer les contours de la fonction
CS = plt.contour(X, Y, Z, levels=20, cmap='viridis')

# Extraire les trajectoires des itérations
trajectory = np.array(trajectory)

# Tracer la trajectoire de la descente de gradient
plt.plot(trajectory[:, 0], trajectory[:, 1], marker='o', color='r', markersize=5, label="Trajectoire")


plt.title("Trajectoire de la descente de gradient")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.colorbar(CS, label="g(x, y)")

# Afficher le graphique
plt.show()
```

On peut observer que la methode de Newton est plus rapide

## Conclusion:

### Avantages et limitations : 
#### Avantages:
- Convergence rapide pour des fonctions convexes ou près d’un minimum.
- Exploite la courbure locale via la matrice hessienne, ce qui permet des mises à jour précises.
- Réduction significative du nombre d’itérations comparée à la descente de gradient.
#### limitations:
- Nécessite le calcul de la matrice hessienne, ce qui est coûteux pour les fonctions à grande dimension.
- Peut échouer en cas de non-convexité si la hessienne est non définie positive ou non inversible.
### Améliorations possibles :
- Utiliser des méthodes quasi-Newton (comme BFGS) pour éviter le calcul exact de la hessienne.
